```python
# Import Libraries
# ---------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Configure pandas
pd.set_option('display.max_columns', None)

# ---------------------------------------------
# Step 2: Define File Paths
# ---------------------------------------------
base_path = r"C:\Users\Administrator\Downloads\Aaaaa_business_folder"
product_path = os.path.join(base_path, "product_sales", "business.retailsales1.csv")
summary_path = os.path.join(base_path, "summary_sales", "business.retailsales2.csv")

# ---------------------------------------------
# Step 3: Load the Datasets
# ---------------------------------------------
df_product = pd.read_csv(product_path)
df_summary = pd.read_csv(summary_path)

```

```python
# ---------------------------------------------
#  Step 4: Basic Overview
# ---------------------------------------------
print("Product Sales Dataset:")
display(df_product.head())
print("\nSummary Dataset:")
display(df_summary.head())

print("\nProduct Info:")
df_product.info()

print("\nSummary Info:")
df_summary.info()

# ---------------------------------------------
#  Step 5: Quick Descriptive Stats
# ---------------------------------------------
print("\nProduct Dataset Stats:")
display(df_product.describe())

print("\nSummary Dataset Stats:")
display(df_summary.describe())

```

```python

#  Clean column names (remove spaces, unify)
df_product.columns = [col.strip().replace(" ", "_") for col in df_product.columns]
df_summary.columns = [col.strip().replace(" ", "_") for col in df_summary.columns]

#Fill missing values properly
# For categorical columns like Product_Type — fill with 'Unknown'
if 'Product_Type' in df_product.columns:
    df_product['Product_Type'] = df_product['Product_Type'].fillna('Unknown')

# For numeric columns — fill missing values with 0
numeric_cols_product = df_product.select_dtypes(include='number').columns
df_product[numeric_cols_product] = df_product[numeric_cols_product].fillna(0)

numeric_cols_summary = df_summary.select_dtypes(include='number').columns
df_summary[numeric_cols_summary] = df_summary[numeric_cols_summary].fillna(0)

```

```python
#  Optional sanity check
print("\nAfter cleaning (product):")
print(df_product.isna().sum())

# ---------------------------------------------
# Step 7: Quick Visual Checks
# ---------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Plot total net sales distribution (product)
plt.figure(figsize=(8,5))
sns.histplot(df_product['Total_Net_Sales'], bins=20, kde=True)
plt.title("Distribution of Total Net Sales (Product Data)")
plt.show()

# Plot gross vs net sales (summary)
plt.figure(figsize=(8,5))
sns.scatterplot(x='Gross_Sales', y='Total_Sales', data=df_summary)
plt.title("Gross Sales vs Total Sales (Summary Data)")
plt.show()
```

```python
# Check that total net sales = gross - discounts - returns (sanity)
df_product["Expected_Net"] = df_product["Gross_Sales"] + df_product["Discounts"] + df_product["Returns"]
inconsistent_rows = df_product[abs(df_product["Expected_Net"] - df_product["Total_Net_Sales"]) > 1]
print(f"\nInconsistent rows in product data: {len(inconsistent_rows)}")
```

```python
# ---------------------------------------------
# Step: Business-Level Duplicate Check
# ---------------------------------------------

# Define the business columns to compare for duplicates
dup_columns = [
    "Product_Type",
    "Net_Quantity",
    "Gross_Sales",
    "Discounts",
    "Returns",
    "Total_Net_Sales",
    "Expected_Net"
]

# Find duplicate rows based on those columns
business_duplicates = df_product[df_product.duplicated(subset=dup_columns, keep=False)]

# Display result summary
print(f"\nPotential business duplicates (same product and values across all key columns): {len(business_duplicates)}")

# If duplicates exist, show a preview
if len(business_duplicates) > 0:
    print(business_duplicates.sort_values(by=dup_columns).head(10))
else:
    print("✅ No business-level duplicates found — data looks consistent!")
```

```python
# ---------------------------------------------
# Compare overall Gross Sales totals
# ---------------------------------------------
total_product_gross = df_product["Gross_Sales"].sum()
total_summary_gross = df_summary["Gross_Sales"].sum()

print(f"Total Gross Sales (Product): {total_product_gross:,.2f}")
print(f"Total Gross Sales (Summary): {total_summary_gross:,.2f}")

difference = total_summary_gross - total_product_gross
ratio = round(total_product_gross / total_summary_gross, 3)

print(f"Difference: {difference:,.2f}")
print(f"Product/Summary Ratio: {ratio}")
```
